```python
!pip install rasterio -q
import rasterio
import numpy as np
from pathlib import Path

# Load the satellite image
tif_path = '/content/drive/MyDrive/xterra_data/carlin_s2.tif'

# Open and read the image
with rasterio.open(tif_path) as src:
    # Read the 4 bands (Red, NIR, SWIR1, SWIR2)
    red = src.read(1)      # Band 4 (Red)
    nir = src.read(2)      # Band 8 (NIR)
    swir1 = src.read(3)    # Band 11 (SWIR1)
    swir2 = src.read(4)    # Band 12 (SWIR2)

    # Get image info
    image_shape = src.shape
    print(f"Image loaded successfully!")
    print(f"Image dimensions: {image_shape}")
    print(f"Bands: Red, NIR, SWIR1, SWIR2")
```

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Convert to float for calculations
red = red.astype('float32')
nir = nir.astype('float32')
swir1 = swir1.astype('float32')
swir2 = swir2.astype('float32')

# Remove invalid values (zeros, negatives)
# Replace zeros with small positive number to avoid division by zero
red = np.where(red <= 0, 0.0001, red)
nir = np.where(nir <= 0, 0.0001, nir)
swir1 = np.where(swir1 <= 0, 0.0001, swir1)
swir2 = np.where(swir2 <= 0, 0.0001, swir2)

print("Calculating mineral indices...")

# 1. IRON OXIDE INDEX
iron_oxide_index = (swir1 - nir) / (swir1 + nir + 1e-8)
iron_oxide_index = np.clip(iron_oxide_index, -1, 1)
iron_oxide_index = np.where(np.isnan(iron_oxide_index), 0, iron_oxide_index)

# 2. CLAY INDEX
clay_index = swir1 / swir2
clay_index = np.clip(clay_index, 0, 2)
clay_index = np.where(np.isnan(clay_index), 0, clay_index)

# 3. SILICA INDEX
silica_index = swir2 / swir1
silica_index = np.clip(silica_index, 0, 2)
silica_index = np.where(np.isnan(silica_index), 0, silica_index)

# 4. K-FELDSPAR INDEX
kfeldspar_index = nir / red
kfeldspar_index = np.clip(kfeldspar_index, 0, 3)
kfeldspar_index = np.where(np.isnan(kfeldspar_index), 0, kfeldspar_index)

# Remove any remaining NaN values
iron_oxide_index = np.nan_to_num(iron_oxide_index, 0)
clay_index = np.nan_to_num(clay_index, 0)
silica_index = np.nan_to_num(silica_index, 0)
kfeldspar_index = np.nan_to_num(kfeldspar_index, 0)

print("âœ… Mineral indices calculated successfully!")
print(f"\nIron Oxide Index:")
print(f"  Mean: {np.nanmean(iron_oxide_index):.3f}")
print(f"  Max: {np.nanmax(iron_oxide_index):.3f}")
print(f"  Min: {np.nanmin(iron_oxide_index):.3f}")

print(f"\nClay Index:")
print(f"  Mean: {np.nanmean(clay_index):.3f}")
print(f"  Max: {np.nanmax(clay_index):.3f}")
print(f"  Min: {np.nanmin(clay_index):.3f}")

print(f"\nSilica Index:")
print(f"  Mean: {np.nanmean(silica_index):.3f}")
print(f"  Max: {np.nanmax(silica_index):.3f}")
print(f"  Min: {np.nanmin(silica_index):.3f}")

print(f"\nK-Feldspar Index:")
print(f"  Mean: {np.nanmean(kfeldspar_index):.3f}")
print(f"  Max: {np.nanmax(kfeldspar_index):.3f}")
print(f"  Min: {np.nanmin(kfeldspar_index):.3f}")
```

```python
print("Classifying lithology (rock types)...")

# Simple lithology classification using band ratios
# GRANITE indicator (quartz + feldspar rich)
granite_index = (swir2 / swir1) * (nir / red)
granite_index = np.clip(granite_index / np.nanmax(granite_index), 0, 1)

# RHYOLITE indicator (silica rich volcanics)
rhyolite_index = swir1 / red
rhyolite_index = np.clip(rhyolite_index / np.nanmax(rhyolite_index), 0, 1)

# BASALT indicator (mafic, iron/magnesium rich)
basalt_index = red / nir
basalt_index = np.clip(basalt_index / np.nanmax(basalt_index), 0, 1)

# Remove NaN
granite_index = np.nan_to_num(granite_index, 0)
rhyolite_index = np.nan_to_num(rhyolite_index, 0)
basalt_index = np.nan_to_num(basalt_index, 0)

print("âœ… Lithology classified!")
print(f"\nGranite Index (ideal host rock for Cu):")
print(f"  Mean: {np.nanmean(granite_index):.3f}")
print(f"  Max: {np.nanmax(granite_index):.3f}")

print(f"\nRhyolite Index (volcanic host):")
print(f"  Mean: {np.nanmean(rhyolite_index):.3f}")
print(f"  Max: {np.nanmax(rhyolite_index):.3f}")

print(f"\nBasalt Index (mafic rocks):")
print(f"  Mean: {np.nanmean(basalt_index):.3f}")
print(f"  Max: {np.nanmax(basalt_index):.3f}")

print(f"\nâœ… Carlin Trend is known as GRANITE-HOSTED")
print(f"Granite Max: {np.nanmax(granite_index):.3f} - Strong signal!")
```

```python
print("Recalculating mineral scores for AOI analysis...")

# Recalculate copper score
copper_score = (
    (kfeldspar_index / 3.0) * 0.4 +
    (clay_index / 2.0) * 0.3 +
    (iron_oxide_index + 1) / 2 * 0.15 +
    (granite_index) * 0.15
)
copper_score_normalized = np.clip(copper_score, 0, 1) * 100

# Recalculate gold score
gold_score = (
    (silica_index / 1.183) * 0.4 +
    (clay_index / 2.0) * 0.3 +
    ((1 - kfeldspar_index/3.0)) * 0.15 +
    (iron_oxide_index + 1) / 2 * 0.15
)
gold_score_normalized = np.clip(gold_score, 0, 1) * 100

print("âœ… Scores recalculated!")
print(f"Copper - Mean: {np.nanmean(copper_score_normalized):.1f}%")
print(f"Gold - Mean: {np.nanmean(gold_score_normalized):.1f}%")
```
